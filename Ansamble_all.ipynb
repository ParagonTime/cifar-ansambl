# CIFAR10

```python
CIFAR_test = torchvision.datasets.CIFAR10('./', download=True, train=False)
```

```python
y_test = torch.LongTensor(CIFAR_test.targets)
```

```python
ResN = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/ansambl_resnet_10.csv')
ResN = ResN.drop(['Unnamed: 0'], axis=1)
```

```python
GoogN = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/ansambl_googlenet_10.csv')
GoogN = GoogN.drop(['Unnamed: 0'], axis=1)
```

```python
VGG16 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/ansambl_vgg16_10.csv')
VGG16 = VGG16.drop(['Unnamed: 0'], axis=1)
```

```python
ADAM = pd.DataFrame()

ADAM['ResNet'] = ResN['Adam']
ADAM['GoogleNet'] = GoogN['Adam']
ADAM['ResNeXt'] = ResNX['Adam']
ADAM['DensN'] = DensN['Adam']
ADAM['VGG16'] = VGG16['Adam']

data = np.zeros((10000,10), dtype=int)
ARes = pd.DataFrame(data)

for i in range(10000):
    for j in range(5):
        if j == 0:
            ARes.iloc[i, ADAM.iloc[i, j]] += 0.8856
        if j == 1:
            ARes.iloc[i, ADAM.iloc[i, j]] += 0.8441
        if j == 2:
            ARes.iloc[i, ADAM.iloc[i, j]] += 0.87
        if j == 3:
            ARes.iloc[i, ADAM.iloc[i, j]] += 0.8749
        if j == 4:
            ARes.iloc[i, ADAM.iloc[i, j]] += 0.8856


anslist = ARes.idxmax(axis=1)
accuracy = (torch.tensor(anslist) == y_test).float().mean().item()
print('Adam:', accuracy)
```
